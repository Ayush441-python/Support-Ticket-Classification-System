```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/customer_support_tickets.csv")
```

```python
df.head()
```

```python
df["text"] = (
    df["Ticket Subject"].astype(str) + " " +
    df["Ticket Description"].astype(str)
)
```

```python
df = df[["text", "Ticket Type", "Ticket Priority"]]
```

```python
df.dropna(subset=["Ticket Type", "Ticket Priority"], inplace=True)
```

```python
df.head()
```

```python
df['Ticket Type'].unique()
```

```python
df["text"].fillna("", inplace=True)
```

```python
df.to_csv("data/processed_data.csv")
```
